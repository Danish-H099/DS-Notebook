```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
plt.plot([3],[4],"ro")
```

```python
plt.plot([1,2,3,4],[1,4,9,16]) 
```

```python
x=np.arange(0,10)
y=np.arange(0,10)
```

```python
plt.plot(x,y)
```

### Graph with label

```python
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.plot(x,y,c='r') # c is color
plt.savefig('Test.png')
```

### Subploting and different styles

```python
x=np.arange(0,13)
y=x*x
```

```python
## Creating Subplots

plt.subplot(2,3,1) # 2rows, 2columns,1st position
plt.plot(x,y,'r--',linewidth=1)

plt.subplot(2,3,2)
plt.plot(x,y,'g*--')

plt.subplot(2,3,3)
plt.plot(x,y,'b*', linestyle="dashed")

plt.subplot(2,3,4)
plt.plot(x,y,'bo')

plt.subplot(2,3,5)
plt.plot(x,y,'g^')

plt.subplot(2,3,6)
plt.plot(x,y,'rs')
```

### Mathematical functions

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("Sin wave graph")
plt.xlabel("angle")
plt.ylabel("value")
plt.plot(x,y)
```

# Bar plot

```python
x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

# Histogram

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

# Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
